```python
sortKey = ["lvy_rper_tin","le_tin","attr_yr_strt_dt"]
sortedData = texData.sort_values(sortKey)
indexData = sortedData.reset_index(drop=True)
indexData
```

```python
indexData = indexData.drop(9)
```

```python
indexData = indexData.reset_index(drop=True)
```

```python
#날짜정보를 계산하기 위한 함수
def dateSetting(inputDate):
    temp = ""
    i = 0
    for i in range(0,3):
        splitValue = inputDate.split("-")[i]
        temp = temp + splitValue
    year = int(temp[0:4])
    month = int(temp[4:6])
    day = int(temp[6:])
    result = datetime(year, month, day)
    return result
```

```python
indexData["continuity_worker"] = ""
```

```python
i = 0
for i in range(i,len(indexData)-1):
    if (indexData.lvy_rper_tin[i] == indexData.lvy_rper_tin[i+1]):
        if(indexData.le_tin[i] == indexData.le_tin[i+1]):
            minusDate = ((dateSetting(indexData.attr_yr_strt_dt[i+1]) - dateSetting(indexData.attr_yr_end_dt[i])).days)
            if(minusDate == 1) :
                indexData["continuity_worker"][i] = "Y"
                indexData["continuity_worker"][i+1] = "Y"
```
