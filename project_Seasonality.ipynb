### 실습3 불량데이터 처리 (QTY 마이너스 값을 0으로 변경한 자료를 QTY_NEW 컬럼에 생성)

```python
myData = pd.read_csv('../lee/Python_ST_EX/dataset/kopo_channel_seasonality_new.csv')
myData["QTY_NEW"] = np.where(myData["QTY"] < 0, 0,\
                             myData["QTY"])
myData.head(5)
```

```python
myData.dtypes
```

```python
# YEAR, WEEK 컬럼 생성하기
```

```python
myData["YEARWEEK"]=myData["YEARWEEK"].astype(str)
```

```python
testData = myData
testData["YEAR"] = testData.YEARWEEK.str[0:4]
testData["WEEK"] = testData.YEARWEEK.str[4:]
testData.head(5)
```

```python
# WEEK가 52이하인 데이터만 추출하기
```

```python
testData.WEEK=testData.WEEK.astype(int)
```

```python
refinedSelloutData = testData[(testData.WEEK <= 52)]
refinedSelloutData.head(5)
```

```python
refinedSelloutData.WEEK.max()
```

### 2-2 데이터 전처리 (이상 데이터 정제하기2) 조건에 따른 값 변경하기 


```python
def refining(df):
    if df["QTY"] < 100:
        return df["QTY"]
    elif df["QTY"] < 1000:
        return 1000
    elif df["QTY"] < 10000:
        return 10000
    else:
        return df["QTY"]
```

```python
# pandas는.. 함수사용 방법이 좀 다른것 같아..
refinedSelloutData["QTY2"] = refinedSelloutData.apply(refining, axis=1)
```

```python
refinedSelloutData.head(5)
```

```python
# 데이터 정렬하기// 처음에는 변수이름을 눈에 보일수 있도록 해주는 것이 좋음.
refinedSelloutData.sort_values(["YEARWEEK","QTY"],ascending=True,inplace=True)
```

```python
refinedSelloutData.head()
```

### 데이터 요약 분석(심화 - 다중집계 <표준편차와 평균 구하기>)
