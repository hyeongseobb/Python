###  2-1. 차트 꾸미기 (선색 및 크기 설정)

```python
xList = list (range(1,11,1))
```

```python
xList
```

```python
y = []
```

```python
# 아래의 포문 디버깅해보기(하나씩 펼쳐보기)
y.append(xList[0]**2)
y
```

```python
for i in xList:
    y.append(i**2)
# .append를 사용하여 xList안에 있는 요소를 앞에서 하나씩 꺼낸값을 2제곱하여 비어있는
# List --> y에 담는다.
```

```python
plt.plot(xList,y, "m*--", lw=1.5)
# 중간에 'm' -> magenta color
#  lw -> 선의 굵기
plt.xlabel('xList') # x축이 뭔지 알도록 써주기
plt.ylabel('y') # y축이 뭔지 있도록 써주기
plt.title('Test Graph') #차트 제목
plt.xlim(4,7) #x의 4-7구간을 확대하여 출력
plt.ylim(0,100)
```

### 2-2. 차트 꾸미기 (범례설정 및 어노테이션)

```python
#레전드 설정
x=list(range(1,11,1))
y = [i**2 for i in x]
#파이썬에서 포문이 간단할 경우 리스트 안에 들어갈 수 있네?
y2 = [i**2.3 for i in x]

plt.plot(x,y,"b*--", label="first")
plt.plot(x,y2,"r*--", label="second")
plt.legend(loc=0) #범례를 만들어줌 (0일시 "best"옵션(빈공간))
plt.annotate('GeniusSeob',xy=(8,70), xytext=(2,4),arrowprops=dict(color="gray", width=1))
#화살표그리고 주석달기
plt.grid(True) #그리드 표기
```

### 3-1. 데이터 시각화 고급(서브플롯 그리기)

```python
x=list(range(1,11,1))
y = [i**2 for i in x]
y2 = [i**2.5 for i in x]
```

```python
fig = plt.figure(figsize=(10,5)) # 도화지를 만들자! 처음에 만들고 싶은 도화지의 크기 (x,y)
ax1 = fig.add_subplot(211) # subplot(2행1렬의1번째)에 서브플롯을 만든다!
plt.plot(x,y, 'b*-', lw=1.5, label = '1st')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('subplot')

ax2 = plt.subplot(212) #subplot(2행 1렬의 2번째)에 서브플롯을 만든다!
plt.plot(x,y2, 'g*-', lw=1.5, label='2nd')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

```python
x=list(range(1,11,1))
y = [i**2 for i in x]
y2 = [i**2.5 for i in x]
y3 = [i**3 for i in x]
```

```python
fig = plt.figure(figsize=(10,5)) # 도화지를 만들자! 처음에 만들고 싶은 도화지의 크기 (x,y)
ax1 = fig.add_subplot(131) # subplot(1행3렬의1번째)에 서브플롯을 만든다!
plt.plot(x,y, 'b*-', lw=1.5, label = '1st')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('subplot')
plt.ylim(0,1000)

ax2 = plt.subplot(132) #subplot(1행 3렬의 2번째)에 서브플롯을 만든다!
plt.plot(x,y2, 'g*-', lw=1.5, label='2nd')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.ylim(0,1000)

ax3 = plt.subplot(133) #subplot(1행 3렬의 3번째)에 서브플롯을 만든다!
plt.plot(x,y3, 'r*-', lw=1.5, label='3rd')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

### 3-2. 데이터 시각화 고급(이중축 활용하기)

```python
x=list(range(1,11,1))
y = [i**2 for i in x]
y2 = [i**2.5 for i in x]
```

```python
fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,1,1) #1행1렬에1개의 서브플롯을 만듬!
plt.plot(x,y, 'bo--',lw=1.5, label = '1st')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('subplot')

ax2 = ax1.twinx() #구분자! twinx() 내가 그림을 차트를 그릴건데 x를 같이 쓰겠다는 의미.
plt.plot(x,y2, 'g', lw=1.5, label='2nd')
plt.plot(x,y2, 'go')
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
```

### 3-3. 데이터 시각화 고급(차트 종류 변경, 차원추가)


```python
#차트 종류 변경
x=list(range(1,11,1))
y = [i**2 for i in x]

plt.figure(figsize=(7,5))

plt.title('Scatter plt')
plt.scatter(x,y,marker='o', edgecolors='b')
```

```python
#차원 추가
x=list(range(1,11,1))
y = [i**2 for i in x]
c= np.random.randint(0,3,len(y))

plt.figure(figsize=(7,5))
plt.scatter(x,y,marker='o',c=c)
plt.colorbar()

plt.title('Scatter plt')
```

### 4. Seaborn

```python
import seaborn as sns

tips = sns.load_dataset('tips')
#웹상에 있는 tips데이터를 불러옴
tips.head(3)
```

```python
tips = sns.load_dataset('tips')
sns.countplot(x="day", data = tips) #countplot -> 특정 컬럼의 개수를 확인할 수 있음.
```

```python
#특정 컬럼 값의 분포도를 확인 하기위한 차트
sns.distplot(tips.total_bill)
```

```python
#2차원 실수형 데이터 시각화
sns.jointplot(x= 'total_bill', y = 'tip', data = tips)
```

```python
#kind = "kde"옵션을 추가하여 커널 밀도함수 시연
sns.jointplot(x= 'total_bill', y = 'tip', data = tips,  kind = "kde")
```

```python
#3차원 데이터 시각화
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue = 'sex')
#hue라는 groupby를 이용함.
```

```python
#barplot 차트 구현
#데이터의 평균 및 실제 범위를 시각화
sns.barplot(x='time', y='total_bill', data = tips)
#데이터의 편차 크기를 선의 길이로 표기함.
```

```python
groupTips = tips.groupby(["time"])["total_bill"].describe()
groupTips2 = groupTips.reset_index()
groupTips2
#위아래의 그래프와 아래의 값을 비교해보자
```

```python
#boxplot 차트 구현
#사분위수 (4등분하는 위치의 수)
#중앙값 : 정렬된 데이터 집합의 중앙에 위치한 값(단, 짝수개인 경우 가운데 두개의 평균값)
#1사분위수: 데이터 하단의 중앙값
#3사분위수: 데이터 상단의 중앙값
# IQR = 3사분위수 - 1사분위수

sns.boxplot(x='time', y='total_bill',data=tips)
```

```python
#데이터의 상관관계를 확인
tips.corr()
```

```python
#자료에 대한 상관관계를 heat map으로 표현
tc = tips.corr()
sns.heatmap(tc,annot = True)
```

```python
#다른 자료 (연도,월별 승객수)
flights = sns.load_dataset('flights')
fp = flights.pivot_table(index='month', columns='year', values = 'passengers')

fp.head()

sns.heatmap(fp)
```

```python
#lmplot
#두 데이터의 상관관계를 linear한 형태로 뽑아내는 차트
sns.lmplot(x='total_bill', y = 'tip', data = tips )
sns.lmplot(x='total_bill', y = 'tip', data = tips, hue = 'sex', markers = ['o','v'] )
#hue는 그룹바이
```
